```python
import matplotlib.pyplot as plt
import pickle
```

```python
list_of_dataset = ['Celegans', 'CSPhD', 'Airport', 'Cora', 'Disease', 'Musae-Twitch']
```

```python
hyp_stats = pickle.load(open('hyperbolicity_stats.pkl', 'rb'))
plt.rcParams.update({'font.size': 25})
plt.figure(figsize = [10,8])
for dataset_name in list_of_dataset:
    if dataset_name != 'Disease':
        plt.hist(hyp_stats[dataset_name]['hyp_avg_list'], bins = 20, density = True, alpha = 0.5, label = dataset_name)
plt.legend(loc = 'upper right')
plt.title('Average Hyperbolicity Distribution')
plt.savefig('AH.pdf', format='pdf')
plt.show()
```

```python
slim_stats = pickle.load(open('slimness_and_other_stats.pkl', 'rb'))
plt.rcParams.update({'font.size': 25})
plt.figure(figsize = [10,8])
for dataset_name in list_of_dataset:
    if dataset_name != 'Disease':
        plt.hist(slim_stats[dataset_name][0], bins = 20, density = True, alpha = 0.5, label = dataset_name)
plt.legend(loc = 'upper right')
plt.title('Average Slimness Distribution')
plt.savefig('AS.pdf', format='pdf')
plt.show()
```
